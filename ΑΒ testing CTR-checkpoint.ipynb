  ### **Parameters of the Model from Power Analysis**
   **-------------------------------------------------------**
  
  
  
  **β:Probability of Type II Error**
  
  **(1-β): Power of the test**
  
  **α: Probability of Type I Error, Significance Level**
  
  **δ: Minimum Detectable Effect**

```python
alpha = 0.05 #5%
print('Alpha: significance level is:', alpha)

delta = 0.1 #10%
print('Delta: minimum detectable effect is:', delta)
```

### Calculating total number of clicks per group by summing clicks

```python
# calculate the total number of records per group

N_con = df_ab_test.groupby('group')['click'].count().loc['con']
N_exp = df_ab_test.groupby('group')['click'].count().loc['exp']

print('Number of records in Control:', N_con)
print('Number of records in Experimental', N_exp)
```

```python
# calculate the total number of clicks per group

X_con = df_ab_test.groupby('group')['click'].sum().loc['con']
X_exp = df_ab_test.groupby('group')['click'].sum().loc['exp']

print('Number of clicks in Control:', X_con)
print('Number of clicks in Experimental', X_exp)
```

### Calculating Pooled Estimates for Clicks per Group

```python
# computing the estimate of click probability per group

p_con = X_con/N_con
p_exp = X_exp/N_exp

print('The probability in Control Group:', p_con)
print('The probability in Experimental Group:', p_exp)

p_pooled = (X_con+X_exp)/(N_con+N_exp)

print('The pooled clicked probability is:', p_pooled)
```

### Calculating Pooled Variance

```python
#computing the estimate of click probability per group

pooled_variance = p_pooled*(1-p_pooled)*(1/N_con+1/N_exp)

print('The pooled variance is:', pooled_variance)
```

### Calculating Standard Error and Test Statistics

```python
# computing the standard error of the test

SE = np.sqrt(pooled_variance)
print('Standard Error is:', SE)

# computing the test statistics of Z-test
Test_stat = (p_con-p_exp)/SE
print('Test Statistics for 2-sample Z-test is:', Test_stat)

#critical value of the Z-test
Z_crit = norm.norm.ppf(1-alpha/2)
print('Z-critical value from standard Normal distribution:', Z_crit)
```

### Calculting p_values of the Z_test


*  A low p-value(p <= 0.05 at 5% significance level) indcated strong evidence against the null hypothesis, so we reject the null hypothesis
*  A high p-value(p > 0.05) indicated weak evidence against the null hypothesis, sp we fail to reject the null hypothesis

```python
# calculate p value

p_value = 2*norm.norm.sf(abs(Test_stat))

# function checking the statistical significance

def is_statistical_significance (p_value, alpha):

    """
    We assess whether there is statistical significance based on the p_value and alpha

    Arguments:
    - p_value (float): The p_value resulting from a statistical test.
    - alpha (float, ortional): The significance level threshold used to determine statistical significance. Defaults to 0.05.

    Returns:
    - Prints the assessment of statistical significance.
    """

# Print the rounded p_value
print(f"P-value of the 2-sample Z-test: {round(p_value, 3)}")

# Determine statistical significance

if p_value <= alpha:
    print( "There is statistical significance, indicating that the observed differences between the groups are unlikely to have occured due to random chance")
else:
    print("There is no statistical significance, suggesting that the observed differences between the groups are likely to have occured")

is_statistical_significance(p_value, alpha)
```

#### So we can reject the original hypothesis Ho: Pcon=Pexp

```python
# Parameters for the standard normal distribution

mu = 0 # Mean
sigma = 1 # Standard deviation

x = np.linspace(mu -3*sigma, mu +3*sigma, 100)
y = norm.norm.pdf(x, mu, sigma)

#Test statistical and critical value from previou calculation
# Test Statistics for 2-sample Z-test is: -59.44163326469381
# Z-critical value from standard Normal distribution: 1.959963984540054

# Plotting the standard distribution
plt.plot(x, y, label='Standard Normal Distribution')

# Shade the rejection region fro a two-tailed test
plt.fill_between(x, y, where=(x > Z_crit) | (x < -Z_crit), color='red', alpha =0.5, label='Regection Region')

# Adding Test Statistic
plt.axvline(Test_stat, color='green', linestyle='dashed', linewidth=2, label=f'Test Statistic = {Test_stat:.2f}')

# Adding Z-Critical values
plt.axvline(Z_crit, color='blue', linestyle='dashed', linewidth=1, label=f'Z-critical = {Z_crit:.2f}')
plt.axvline(-Z_crit, color='blue', linestyle='dashed', linewidth=1)

#Adding labels and title
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.title('Gaussian Distributionwith Rejection egion \n (A/B Teting for CTR button)')
plt.legend()

plt.show()
```

```python
# Calculate the Confidence Interval (CI) for a 2-sample Z-Test
# Calculate the lower and upper bounds of the confidence interval

CI = [
    round((p_exp - p_con) - SE*Z_crit, 3), #lower bound of the CI
    round((p_exp - p_con) + SE*Z_crit, 3) #upper bound of the CI
]

print('Confidence Interval of the 2 sample Z-Test is:', CI)
```

#### We see that the Confidence Interval is pretty narrow so our results are accurate and can be generilized with safety

```python
def is_practicaly_significant(delta, CI_95):
    
    """
    We assess here if the difference between Control andExperimental Group is practically significant using MinimumDetectable Effect (MDE)

    Arguments:
    - delta (float): The minimum Detectable Efffect (MDE) considered for practical significance.
    - CI_95 (tuple): A tuple representing the lower and upper bounds of the 95% Confidence level.

    Returns:
    - Prints whether the experiment has practical significance based on th e MDE and CI
    """
    
    # Extracts lower bound of 95% CI
    lower_bound_CI = CI_95[0]
    
    # Check if the lower bound of the CI is greater than or equal to delta
    if lower_bound_CI <= delta:
        print(f"We have practical significane! \nWith MDE of {delta}, The difference between Control and Experimental group is practically significant")
        return True
    else:
        print (f"We don't have practical significane! \nThe difference between Control and Experimental group is not practically significant")
        return False
        
# Example of using the function
delta = 0.05 # Assuming 5% increase is our MDE
CI_95 = (0.04, 0.06) # Example Confidence Interval

significance = is_practicaly_significant(delta, CI_95)
print('Lower bound of 95% confidence interval is:', CI_95[0])
```
